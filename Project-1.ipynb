```python
data_set = pd.read_csv('Laliga.csv')
data_set.head()
```

```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

data_set.replace(to_replace='-',value=0,inplace=True)
data_set.sample(10)
```

```python
# Print all the teams which have started playing between 1930-1980.

data_set[data_set.Debut.apply(lambda x:(int(x[0:4])>=1930 and int(x[0:4])<=1980))][['Team','Debut']]
```

```python
# Print the list of teams which came Top 5 in terms of points

data_set.astype({'Points':int},copy=False).nlargest(5,'Points')[['Team','Points']]
```

```python
# Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.
# Goal_diff_count = GoalsFor - GoalsAgainst

def Goal_diff_count(GoalsFor,GoalsAgainst):
    return(GoalsFor-GoalsAgainst)

print('Teams with goal difference')
new_df=data_set.assign(Goal_diff=data_set.GoalsFor.astype(int).pipe(Goal_diff_count,data_set.GoalsAgainst.astype(int)))[['Team','Goal_diff']]
print(new_df)
print()
print('Maximum goal difference team')
print(new_df.nlargest(1,['Goal_diff']))
print()
print('Minimum goal difference team')
print(new_df.nsmallest(1,['Goal_diff']))
```

```python
#Create a new column with name “Winning Percent” and append it to the data set
#Percentage of Winning = (GamesWon / GamesPlayed)*100
#If there are any numerical error, replace it with 0%
#Print the top 5 teams which has the highest Winning percentage

data_set=data_set.assign(Winning_Percent=lambda x :(x.GamesWon.astype(int)/x.GamesPlayed.astype(int))*100).fillna(0)
data_set.nlargest(5,['Winning_Percent'])[['Team','Winning_Percent']]
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions

data_set[['BestPosition','Points']].astype({'Points':int}).groupby('BestPosition',axis=0).sum()
```
