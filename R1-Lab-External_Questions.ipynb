### Plot the attrition column (Since it is a categorical variable, chose the plot accordingly)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
yes_cnt = hr[hr['Attrition']=='Yes'].count()['Attrition']
no_cnt = hr[hr['Attrition']=='No'].count()['Attrition']
#plt.figure(figsize=(1,1))
plt.ylabel('No. of Employees')
plt.xlabel('Attrition')
plt.bar(x=['Yes','No'],height=[yes_cnt,no_cnt],width=0.2)
```

### Plot Total working years using matplotlib plot function

```python
minimum = hr['TotalWorkingYears'].min()
maximum = hr['TotalWorkingYears'].max()
val_range = maximum-minimum
bin_size= val_range/8
bins = np.arange(minimum,maximum+bin_size,bin_size,dtype=int)
plt.xticks(bins)
plt.xlabel('Total Working Years')
plt.ylabel('No. of Employees')
plt.hist(x=hr['TotalWorkingYears'],bins=bins)
```
